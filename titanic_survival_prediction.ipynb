```python
# Step 1: Import Libraries and Initial Setup
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# make plots look nice
plt.style.use('fivethirtyeight')   
# show plots inside notebook
%matplotlib inline   
# hide warning messages              
warnings.filterwarnings('ignore')  
```

```python

train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')

# To know the number of columns and rows
train.shape 
```

```python
# To know the info about each column like data type, ect
train.info()
```

```python
# Check for any NULL values
train.isnull().sum()
```

```python
# 2. Data Visualisation

# 2.1 Number of survivors and death count
f, ax = plt.subplots(1, 2, figsize=(12, 4))

# Pie Chart
train['Survived'].value_counts().plot.pie(
    explode=[0, 0.1],
    autopct='%1.1f%%',
    ax=ax[0],
    shadow=False
)
ax[0].set_title('Survivors (1) and the Dead (0)')
ax[0].set_ylabel('')

# Bar chart
sns.countplot(x='Survived', data=train, ax=ax[1])
ax[1].set_title('Survivors (1) and the dead (0)')

plt.show()
```

```python
# 2.2 Analysing the Impact of Sex on Survival Rates

f, ax = plt.subplots(1, 2, figsize=(12, 4)) 

# Bar plot showing average survival rate by sex
train[['Sex', 'Survived']].groupby(['Sex']).mean().plot.bar(ax=ax[0]) 
ax[0].set_title('Survival Rate by Sex')
ax[0].set_ylabel('Survival Rate')

# Count plot showing number of survivors and non-survivors by sex
sns.countplot(x='Sex', hue='Survived', data=train, ax=ax[1])
ax[1].set_ylabel('Quantity') 
ax[1].set_title('Survived (1) and deceased (0): men and women') 

plt.show()
```
